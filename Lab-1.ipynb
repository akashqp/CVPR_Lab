```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

**Reading, Displaying and writing an image using OpenCV**

```python
image = cv.imread("scream.jpg")
```

```python
image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
```

```python
plt.axis('off')
plt.imshow(image)
```

```python
cv.imwrite("output_img.jpeg", image)
```

**Convert image to another format**

```python
cv.imwrite("output_img.png", image)
```

**Perform image resizing**

```python
resize_img = cv.resize(image, (150, 150))
plt.axis('off')
plt.imshow(resize_img)
```

**Convert coloured image into grayscale image**

```python
gray_img = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap='gray')
plt.axis('off')  # Hide axes
plt.show()
```

**Scaling, rotation and shifting operations**

```python
scaled_img = cv.resize(image, None, fx=0.5, fy=0.5)
plt.imshow(scaled_img)
plt.axis('off')  # Hide axes
plt.show()
```

```python
rows, cols, channels = image.shape
rotation_matrix = cv.getRotationMatrix2D((cols / 2, rows / 2), 90, 1)
rotated_img = cv.warpAffine(image, rotation_matrix, (cols, rows))
plt.imshow(rotated_img)
plt.axis('off')  # Hide axes
plt.show()
```

```python
translation_matrix = np.float32([[1, 0, 10], [0, 1, 40]])
shifted_img = cv.warpAffine(image, translation_matrix, (cols, rows))
plt.imshow(shifted_img)
plt.axis('off')  # Hide axes
plt.show()
```
