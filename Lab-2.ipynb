# **Lab-2:**

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv.imread("sun_flower.jpg")
image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image = cv.resize(image, None, fx=0.5, fy=0.5)
```

```python
plt.imshow(image)
plt.axis('off')  # Hide axes
plt.show()
```

**Convert image into grayscale image**

```python
gray_img = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap='gray')
plt.axis('off')  # Hide axes
plt.show()
```

**Convert image to binary**

```python
_, binary_img = cv.threshold(gray_img, 128, 255, cv.THRESH_BINARY)

# Display the binary image
plt.imshow(binary_img, cmap='gray')
plt.axis('off')
plt.show()
```

**Split image into RGB components**

```python
R, G, B = cv.split(image)
```

```python
plt.imshow(R)
plt.title('Red Channel')
plt.axis('off')
```

```python
plt.imshow(G)
plt.title('Green Channel')
plt.axis('off')
```

```python
plt.imshow(B)
plt.title('Blue Channel')
plt.axis('off')
```

**Merge RGB componenets**

```python
# Merge the RGB channels
merged_image = cv.merge((R, G, B))

# Display the merged image
plt.imshow(merged_image)
plt.axis('off')
plt.show()
```
