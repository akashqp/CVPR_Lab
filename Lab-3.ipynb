```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv.imread("sun_flower.jpg", cv.IMREAD_GRAYSCALE)
image = cv.resize(image, None, fx=0.5, fy=0.5)
```

**Contrast Stretching**

```python
min_intensity = np.min(image)
max_intensity = np.max(image)
stretched_img = 255 * ((image - min_intensity) / (max_intensity - min_intensity))
stretched_img = stretched_img.astype(np.uint8)
```

```python
plt.imshow(stretched_img, cmap='gray')
plt.axis('off')
plt.show()
```

**Histogram Equilization**

```python
equalized_image = cv.equalizeHist(image)
```

```python
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')
```

**Gamma Correction**

```python
gamma = 1.5
gamma_corr = np.uint8(cv.pow(image / 255.0, gamma) * 255)
plt.imshow(gamma_corr, cmap='gray')
plt.title('Gamma Corrected Image (gamma={})'.format(gamma))
plt.axis('off')
```

**Thresholding**

```python
threshold_val = 145
_, threshold_img = cv.threshold(image, threshold_val, 255, cv.THRESH_BINARY)
plt.imshow(threshold_img, cmap='gray')
plt.title('Thresholded Image (Threshold={})'.format(threshold_val))
plt.axis('off')
plt.show()
```

**Sigmoidal Contrast Enhancement**

```python
# Define parameters for sigmoidal contrast enhancement
beta = 10.0
mid_intensity = np.mean(image)

# Perform sigmoidal contrast enhancement
sigmoidal_image = 1 / (1 + np.exp(beta * (mid_intensity - image) / 255.0))

# Scale the enhanced image back to the 0-255 range
sigmoidal_image = (sigmoidal_image * 255).astype(np.uint8)
plt.imshow(sigmoidal_image, cmap='gray')
plt.title('Sigmoidal Contrast Enhanced Image')
plt.axis('off')
plt.show()
```
