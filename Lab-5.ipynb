```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = plt.imread('sunflower.jpg')
```

```python
X = image.reshape((-1, 3))
```

## **Principal Component Analysis**

**Create a PCA instance**

```python
pca = PCA(n_components=0.95)  # Preserve 95% of the variance
```

```python
X_pca = pca.fit_transform(X)
```

**Reconstruct the image from the reduced PCA components**

```python
X_reconstructed = pca.inverse_transform(X_pca).reshape(image.shape)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original Image')
plt.subplot(1, 2, 2)
plt.imshow(X_reconstructed, cmap='gray')
plt.title('Reconstructed Image (PCA)')
plt.show()
```
