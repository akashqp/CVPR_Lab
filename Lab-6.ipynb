```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.segmentation import active_contour
from skimage.io import imread
```

```python
image = imread('sunflower.jpg', as_gray=True)
```

## **Active Contour**

**Initialize the snake contour**

```python
s = np.linspace(0, 2*np.pi, 100)
x = 100 + 100*np.cos(s)
y = 100 + 100*np.sin(s)
init = np.array([x, y]).T
```

**Apply active contour algorithm**

```python
snake = active_contour(image, init, alpha=0.015, beta=10, gamma=0.001)
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(image, cmap='gray')
plt.plot(init[:, 0], init[:, 1], '--r', lw=3)
plt.plot(snake[:, 0], snake[:, 1], '-b', lw=3)
plt.axis('off')
plt.title('Active Contour Segmentation')
plt.show()
```
